```python
import pandas as pd
import os
```

```python
path  = "yelp_review.csv"
df = pd.read_csv(path)
```

```python
sample_df = df.sample(frac=0.1, replace=True, random_state=1)
```

```python
sample_df.count()
```

```python
one_star = sample_df.loc[sample_df['stars'] == 1]
one_star.head()
```

```python
one_star = one_star.sample(n=1000, replace=False, random_state=1)
one_star.head()
```

```python
two_star = sample_df.loc[sample_df['stars'] == 2]
two_star.head()
```

```python
two_star = two_star.sample(n=1000, replace=False, random_state=1)
two_star.head()
```

```python
two_star.count()
```

```python
three_star = sample_df.loc[sample_df['stars'] == 3]
three_star.head()
```

```python
three_star = three_star.sample(n=1000, replace=False, random_state=1)
three_star.head()
```

```python
three_star.count()
```

```python
four_star = sample_df.loc[sample_df['stars'] == 4]
four_star.head()
```

```python
four_star = four_star.sample(n=1000, replace=False, random_state=1)
four_star.head()
```

```python
four_star.count()
```

```python
five_star = sample_df.loc[sample_df['stars'] == 5]
five_star.head()
```

```python
five_star = five_star.sample(n=1000, replace=False, random_state=1)
five_star.head()
```

```python
five_star.count()
```

```python
df_name_list = [one_star, two_star, three_star, four_star, five_star,]
uniform_df = pd.concat(df_name_list)
uniform_df.head()
```

```python
uniform_df = uniform_df.sample(frac=1).reset_index(drop=True)
```

```python
uniform_df['stars'].value_counts()
```

```python
file_name = 'uniform_yelp.csv'
uniform_df.to_csv(file_name, sep=',', encoding='utf-8',index=False)
```
